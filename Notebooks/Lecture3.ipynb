$$
\begin{aligned}
&u_t + f (u)_x = 0, \quad[\text{differential form}]\\
&u(x,0)=u^0(x).
\end{aligned}
$$

The concept of weak solution can be introduced by considering the integral form:
$$
\frac{d}{dt}
\int^b_a  u(x, t) dx=f(u(a, t))-f(u(b,t)) \quad[\text{intergral form}]
$$

For $C^1$ solutions, the solution defined by strong form and weak form are equivalent. Attempts at defining weak solutions:

- If u satisfies [intergral form] for almost all $(a,b)$ then in $u$ is called a weak solution to the differential form. (physically meaningful,correct)
- If for any $\phi\in C^1_0(\mathbb{R}^2)$,
$$ - \int_0^\infty\int_0^\infty (u\phi_t+f(u)\phi_x) dxdt- \int_{-\infty}^\infty u^0(x)\phi(x,0)dx=0,$$
&emsp;&emsp;then $u$ is called a weak solution to [differential form]. (more meaningful mathematically motivated by multiplication by test function and integration by parts.)

It turns out the two are equivalent. (Not proven here.) Now, assume a solution that has two $C^1$ segments separated by a curve on which no regularity is demanded of $u$, $x(t)$ is the function of the discontinuity position, then 
$$
\begin{aligned}
0 = & \frac{d}{dt}\int^b_a u(x,t)dx+f(u(b,t))-f(u(a,t)) \\
  = & \frac{d}{dt}\Big[\int^{x(t)}_a u(x,t)dx+\int^{b}_{x(t)} u(x,t)dx\Big]+f(u(b,t))-f(u(a,t)) \\
  = & u(x(t^-),t)x'(t) + \int^{x(t)}_a u_t(x,t)dx - u(x(t^+), t)x'(t)+\int^{b}_{x(t)} u_t(x,t)dx + f(u(b,t))-f(u(a,t)) \\
  = & u(x(t^-),t)x'(t) - \int^{x(t)}_a f(u)_xdx - u(x(t^+), t)x'(t)-\int^{b}_{x(t)} f(u)_x dx + f(u(b,t))-f(u(a,t)) \\
  = & u(x(t^-),t)x'(t) - f(u(x(t^-),t)) + f(u(a,t)) - u(x(t^+), t)x'(t)  - f(u(b,t)) + f(u(x(t^+),t))\\
    & + f(u(b,t))-f(u(a,t)) \\
  = & u(x(t^-),t)x'(t) - f(u(x(t^-),t)) + \sout{f(u(a,t))} - u(x(t^+), t)x'(t)  - \sout{f(u(b,t))} + f(u(x(t^+),t))\\
    & + \sout{f(u(b,t))}-\sout{f(u(a,t))} \\
  = & u(x(t^-),t)x'(t) - f(u(x(t^-),t))  - u(x(t^+), t)x'(t)  + f(u(x(t^+),t))
  \end{aligned}
$$

Use the shorthand
$$
u^+ := u(x(t^+),t)), \quad u^- := u(x(t^-),t))
$$
and write 
$$
0 = f(u^+) - f(u^-) - x'(t)(u^+-u^-).
$$
If $u^+\neq u^-$, we get the _Rankine-Hugoniot jump condition_:
$$
    x'(t) = \frac{f(u^+)-f(u^-)}{u^+- u^-}
$$
If u is piecewise $C^1$ and is discontinuous only along isoated curves, and if u satisfies the PDE when it is $C^1$, and the Rankine-Hugoniot (RH) condition along all discontinuous cruves, then u is a weak solution.


# Entropy Solution
We need a third category of solutions, called entropy solutions, where neither uniqueness nor existence poses a big problem. Consider adding an artificial viscosity:
$$ u^\epsilon_t + f(u^\epsilon)_x = \epsilon u ^\epsilon_{xx}$$
with a very small $0<\epsilon\ll 1$.
Then we would wish to define an entropy solution as
$$\lim_{\epsilon\to 0} u^\epsilon(x,t) = u(x,t)
$$
in some norm. In fact, this is _the_ entropy solution. 


Pick a function $U(u)$ called the _entropy function_ if $U''(u)\geq 0$, i.e. if it is convex. Then multiply the conservation law with viscosity by $U'(u^\epsilon)$:
$$
\begin{aligned}
    U'(u^\epsilon)(u^\epsilon_t + f(u^\epsilon)_x) &= \epsilon U'(u^\epsilon)u_{xx}^\epsilon\\
    U(u^\epsilon)_t + F(u^\epsilon)_x &= \epsilon [(U'(u^\epsilon)u_{x}^\epsilon)_x - U''(u^\epsilon)(u^\epsilon_x)^2]\\
    U(u^\epsilon)_t + F(u^\epsilon)_x &\leq  \epsilon (U'(u^\epsilon)u_{x}^\epsilon)_x 
\end{aligned}
$$
where 
$$ 
F(u) = \int^u U'(v)f'(v)dv \Rightarrow F'(u) = U'(u)f'(u).
$$
To support our argument as $\epsilon\to 0$, once again take a test function $\phi\in C^2_0(\mathbb{R}\times\mathbb{R}^+)$, $\phi\geq 0$.
$$
\begin{aligned}
    \int^\infty_0\int^\infty_{-\infty}(U'(u^\epsilon)_t + F(u^\epsilon)_x)\phi dxdt &\leq  \epsilon\int^\infty_0\int^\infty_{-\infty}(U'(u^\epsilon)u_{x}^\epsilon)_x \phi dx dt\\
    \Rightarrow\int^\infty_0\int^\infty_{-\infty}U(u^\epsilon)\phi_t + F(u^\epsilon)\phi_x dxdt &\geq  \epsilon\int^\infty_0\int^\infty_{-\infty}U'(u^\epsilon)u_{x}^\epsilon \phi_x dx dt   \\
    & = -\epsilon\int^\infty_0\int^\infty_{-\infty}U(u^\epsilon) \phi_{xx} dx dt 
\end{aligned}
$$

DCT allows taking the limit, we get the _entropy inequality_
$$\int^\infty_0\int^\infty_{-\infty}U(u)\phi_t + F(u)\phi_x dxdt \geq  0 $$

**Definition**. A conservation law is called genuinely nonlinear iff $f''(u) \neq 0$. If $f''(u) > 0$, it is called convex, if $f''(u) < 0$ it is called concave.

Shocks must appear for genuinely nonlinear conservation laws under periodic or compactly supported initial conditions.

Consider a box containing the support of a test function $\phi\in C_{c}^\infty(\mathbb{R}\times \mathbb{R}^+)$ and let $u(x,t)$ be piecewise $C^1$ with on the discontinuity along $(t,x(t))$.

Then consider 
$$ 
\begin{aligned}
0\geq & - \int_{c}^d\int_a^b (U(u)\phi_t + F(u)\phi_x)dxdt\\
 = &  - \int_{c}^d\int_a^{x(t)} (U(u)\phi_t + F(u)\phi_x)dxdt  - \int_{c}^d\int_{x(t)}^b (U(u)\phi_t + F(u)\phi_x)dxdt\\
 = & \int_{c}^d\int_a^{x(t)} \underbrace{(U(u)_t + F(u)_x)}_{=0}\phi dxdt +  \int_{c}^d\int_{x(t)}^b \underbrace{(U(u)_t + F(u)_x)}_{=0}\phi dxdt \\
  & - \int_{\partial\Omega^-} \phi(U(u),F(u))\cdot \mathbf{n} ds - \int_{\partial\Omega^+} \phi(U(u),F(u))\cdot \mathbf{n} ds \\
= &  \int_{\Gamma} \phi\frac{x'(t)U(u^-)-F(u^-)}{\sqrt{1+(x'(t))^2}} ds - \int_{\Gamma} \phi\frac{x'(t)U(u^+)-F(u^+)}{\sqrt{1+(x'(t))^2}} ds \\
= &  \int_{\Gamma} \frac{\phi}{\sqrt{1+(x'(t))^2}}[x'(t)(U(u^-)-U(u^+))-(F(u^-)-F(u^+))] ds
\end{aligned}
$$

We have 
$$
x'(t)(U(u^-)-U(u^+))-(F(u^-)-F(u^+))\leq 0.
$$

Introduce the notation $\llbracket f\rrbracket: = f(u^+)- f(u^-)$, then this condition becomes
$$ x'(t) \llbracket U\rrbracket \geq \llbracket F\rrbracket

Oleinik entropy condition: For all $u$ between $u^-$ and $u^+$, we need to have 
$$
\frac{f(u)-f(u^-)}{u-u^-} \geq x'(t) \geq \frac{f(u)-f(u^+)}{u-u^+} 
$$
where $s=x'(t)$ is the shock speed, known from the R-H condition.


Lax Entropy Condition:
$$ f'(u^-) > s > f'(u^+). $$

**Exercise** Show which solution is entropy solution. 

It is easy to see that the Oleinik condition implies Lax's condition. Unfortunately, the converse does not hold. Lax's entropy condition does not guarantee uniqueness but it is a necessary condition. However, if $f''(u) > 0$ or $f''(u)<0$ uniformly (i.e. the conservation law is genuinely nonlinear), then Lax's entropy condition is sufficient for $u$ to be the entropy solution.

For $f'(u)>0$, Lax's condition becomes even simpler. Consider
$$ 
f'(u^-) \geq s = \frac{\llbracket f(u)\rrbracket}{\llbracket u\rrbracket}\geq f'(u^+).
$$
and note that $f'(u)$ is monotonically increasing, such that the middle part is automatically satisfied. Thus, Lax's condition becomes
$$
f'(u^-)\geq f'(u^+)
$$
i.e., looking towards the right, we can only jump down.

**Theorem** The solutions to 
$$
\left\{
\begin{array}{l}
    u_t^\epsilon + f(u^\epsilon)_x = \epsilon u^\epsilon_{xx}\\
    u^\epsilon(x,0) = u^0(x)
\end{array}
\right.
$$
are $L^1$ contractive. Namely, let $v^\epsilon$ be the solution of 
$$
\left\{
\begin{array}{l}
    v_t^\epsilon + f(v^\epsilon)_x = \epsilon v^\epsilon_{xx}\\
    v^\epsilon(x,0) = v^0(x)
\end{array}
\right.
$$
Then 
$$
\| u^\epsilon(\cdot, t)-v^\epsilon(\cdot, t)\|_{L^1}\leq \|u^0-v^0\|_{L^1}.
$$

**Proof**

We need to show that 
$$ 0 \geq \frac{d}{dt} \int^\infty_{-\infty} |u^\epsilon(x,t)-v^\epsilon(x,t|dx.$$

Let $s_j$ be the sign of $u^\epsilon - v^\epsilon$ on $I_j$, $|u^\epsilon-v^\epsilon| = s_j (u^\epsilon-v^\epsilon) $, and consider, using Leibniz's rule, the following 
$$
\begin{aligned}
 & \frac{d}{dt} \int^\infty_{-\infty} |u^\epsilon(x,t)-v^\epsilon(x,t)|dx\\
= &  \frac{d}{dt} \sum_j \int^{x_{j+1/2}(t)}_{x_{j-1/2}(t)} s_j(t)(u^\epsilon(x,t)-v^\epsilon(x,t))dx\\
= & \sum_j s_j(t)\Big[ \underbrace{u^\epsilon(x_{j+1/2}(t),t)-v^\epsilon(x_{j+1/2}(t),t)}_{0}\Big]x'_{j+1/2}(t)\\
& - s_j(t)\Big[ \underbrace{u^\epsilon(x_{j-1/2}(t),t)-v^\epsilon(x_{j-1/2}(t),t)}_{0}\Big]x'_{j-1/2}(t)\\
& + \int^{x_{j+1/2}}_{x_{j-1/2}} \underbrace{s'_{j}(t)}_{0}(u^\epsilon(x,t)-v^\epsilon(x,t))dx\\
& + \int^{x_{j+1/2}}_{x_{j-1/2}} s_{j}(t)(u_t^\epsilon(x,t)-v_t^\epsilon(x,t))dx\\
 = & \sum_j \int^{x_{j+1/2}}_{x_{j-1/2}} s_{j}(t)(u_t^\epsilon(x,t)-v_t^\epsilon(x,t))dx\\
 = & \sum_j \int^{x_{j+1/2}}_{x_{j-1/2}} s_j(t)[-f(u^\epsilon)_x + \epsilon u^\epsilon_{xx}+f(v^\epsilon)_x - \epsilon v^\epsilon_{xx}]dx\\
 = & \sum_j s_j(t)\Big\{ \underbrace{-f(u^\epsilon(x_{j+1/2}(t),t))+f(u^\epsilon(x_{j-1/2}(t),t))+f(v^\epsilon(x_{j+1/2}(t),t))-f(v^\epsilon(x_{j-1/2}(t),t))}_{=0}  \\
&  + \epsilon[ {\color{red} u_x^\epsilon(x_{j+1/2}(t),t)} - {\color{blue} u_x^\epsilon(x_{j-1/2}(t),t)} - {\color{red}v_x^\epsilon(x_{j+1/2}(t),t)} + {\color{blue} v_x^\epsilon(x_{j-1/2}(t),t)} ] \Big\}\\
& \leq 0
\end{aligned}
$$
By definition of $x_{j\pm 1/2}$ and the sign of $s_j(t)$, we have the results. 


The entropy solution has a non-increasing total variation.

$$ \mathrm{TV} (u(\cdot, t))\leq \mathrm{TV}(u^0).$$
